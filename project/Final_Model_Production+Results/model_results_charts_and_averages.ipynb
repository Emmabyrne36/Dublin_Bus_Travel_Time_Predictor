## Gradient Boosting Regression (loss = 'huber') - no outliers cut
### Note: This is the model used in the application

```python
gbr_noc = pd.read_csv('model_results_GBR_no_outliers_cut.csv')
gbr_noc.shape[0]
```

```python
gbr_noc = gbr_noc.drop_duplicates()
gbr_noc.shape[0]
```

```python
gbr_noc.sort_values(['MdAPE', 'MAPE', 'R-squared'], inplace=True)
gbr_noc
```

```python
print ("Average MdAPE is", gbr_noc.MdAPE.mean())
print ("Average MAPE is", gbr_noc.MAPE.mean())
print ("Average R-squared value is", gbr_noc['R-squared'].mean())
```

#### Note: In the following step all SSIDs of fewer than 25 entries were removed from consideration for final calculations/graphing; this was due to the likelihood of these being erroneous in some way, as it is very unlikely that any segment would have fewer than 4 buses a week traversing them.  We did not cut segments with any more entries than this, as some JourneyPatternIDs do refer to buses that run a very small number of times a week, sometimes only one day a week.  All SSID pickle files have been kept on the server however, in case some of these were actually valid and the models required by the app.  The scale of the data (over 5,500 segments represented) meant that we were unable to verify the validity of each and every SSID.  Thi step is repeated for all models tested below.

```python
gbr_noc['Underreported'] = gbr_noc['Row_Count'] < 25
gbr_noc = gbr_noc[gbr_noc.Underreported != True]
gbr_noc = gbr_noc.drop(['Underreported'], axis=1)
gbr_noc.reset_index(inplace=True)
gbr_noc = gbr_noc.drop('index', axis=1)
gbr_noc
```

```python
print ("Average MdAPE is", gbr_noc.MdAPE.mean())
print ("Average MAPE is", gbr_noc.MAPE.mean())
print ("Average R-squared value is", gbr_noc['R-squared'].mean())
```

```python
fig = plt.figure(figsize=(16,10))
plt.rc('font', size=16)
ax = fig.add_subplot(111)

ax.plot(gbr_noc.MAPE, label='MAPE', color='c')
ax.plot(gbr_noc.MdAPE, label='MdAPE', color='r')

ax.set_title('GBR Median and Mean Absolute Percentage Error per Segment \n (no outliers cut)')
ax.legend(loc='upper left')
ax.set_ylabel('Percentage')
ax.set_xlabel('Segments')
```
