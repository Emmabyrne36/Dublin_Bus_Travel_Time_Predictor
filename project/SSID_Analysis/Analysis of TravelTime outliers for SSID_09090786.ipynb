```python
import pandas as pd
import numpy as np
```

```python
# Examining 2 second TravelTime value for SSID09090786
df = pd.read_csv('siri.20121127.csv', header=None)
pd.set_option('display.max_rows', 100)
```

```python
# rename iterative numeric header row to column names
df = df.rename(columns={0:'Timestamp', 1:'LineID', 2:'Direction', 3:'JourneyPatternID', 4:'TimeFrame', 5:'VehicleJourneyID', 6:'Operator', 7:'Congestion', 8:'Lon', 9:'Lat', 10:'Delay', 11:'BlockID', 12:'VehicleID', 13:'StopID', 14:'AtStop'}) 
```

```python
# drop some unnecessary rows so the table fits on one page
df = df.drop(['Direction', 'Congestion', 'Operator', 'BlockID', 'VehicleID'], axis=1)
```

```python
df.dtypes
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='us')
```

```python
# only look at the relevant VehicleJourneyID
timesample = df[df.VehicleJourneyID == 6373]
```

```python
# only look at the relevant StopIDs - the ones for the SSID in question, and the one before.
timesample = timesample[(timesample.StopID == '908') | (timesample.StopID == '909') | (timesample.StopID == '786')]
```

```python
timesample
```

The gap between StopID909(just one entry, AtStop=0) and StopID786(first entry, AtStop=1) is actually 2 seconds in the original data - no errors here our end.

```python
del df
del timesample
```
