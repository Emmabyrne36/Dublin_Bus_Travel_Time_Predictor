```python
# Inserting 'null' string as replacement for missing values in JourneyPatternID, StopID, LineID
res['JourneyPatternID'] = res['JourneyPatternID'].fillna('null')
res['StopID'] = res['StopID'].fillna('null')
```

```python
#O

res.info(memory_usage='deep')
```

```python
#O

# Checking for missing (null) values
res.isnull().sum()
```

```python
# Converting 'Timestamp' to date-time format
res['Timestamp'] = pd.to_datetime(res['Timestamp'], unit='us')
```

```python
#O

res.info(memory_usage='deep')
```

```python
# Converting 'TimeFrame' to day-of-the-week feature
res['TimeFrame'] = pd.to_datetime(res['TimeFrame'], format='%Y-%m-%d')
res['Day'] = res['TimeFrame'].dt.weekday_name

# Commenting out dropping the TimeFrame column in case it's needed as part of the primary key (as Wen-Ting indicated it might)
#res = res.drop('TimeFrame', axis=1)
```

```python
#O

res.info(memory_usage='deep')
```

```python
# Removing all rows where JourneyPatternID = 'null' 

res = res[res.JourneyPatternID != 'null']
```

```python
#O

res.info(memory_usage='deep')
```

```python
#O

# The number of rows matches up with total rows minus null JourneyPatternID rows, so this has worked properly.
res.shape
```

```python
#O

# Looking at the 500+ remaining StopID = 'null' rows

pd.set_option('display.max_rows', 2000)
res[res.StopID == 'null']
```

```python
#O

# Now looking at all rows for VehicleID=40025 & TimeFrame=2013-01-30, as an example of StopID = 'null'
# These two 'nulls' are the 5th and 6th entries in the table above (row nos. 1184935 and 1184987)
#
# We can see below that these two StopID = 'null' values occur at the end of a length of time where JourneyPatternID=040B0001
# is sitting at a stop, before switching to 040D0002, doing this route, and then switching to 040D1001 and doing that.
# The only time during either of these journeys that StopID=1 is at the terminus!
# I don't know if this means no one got on the bus or what, but it seems anomolous.
#
# Note also that BlockID remains the same throughout all these journeys - 40B and two directions of 40D
# The last three digits may vary - we'll check this shortly, but first...

res[(res.TimeFrame == '2013-01-30') & (res.VehicleID == 40025)]
```

```python
#O

# ...we'll check if any values for this BlockID exist that are not '40' buses.
# And no, they don't.

res[(res.BlockID == 40208) & (~res['JourneyPatternID'].str.contains("40"))]
```

```python
#O

# Now we'll look at all BlockIDs that start with 40 and see if any do not correspond to '40' buses
# And we see that a very small number do not, so these are presumably errors

block40 = res[(res.BlockID >= 40000) & (res.BlockID < 41000)]
block40.groupby(['JourneyPatternID']).size()
```

```python
#O

# Now we'll do the reverse, and look at all JourneyPatternIDs that start with 40 (excluding the '4' and '104')
# and see if any do not correspond to 40xxx BlockIDs

jpid40 = res[(res['JourneyPatternID'].str.contains("040")) & (~res['JourneyPatternID'].str.contains("00040")) & (~res['JourneyPatternID'].str.contains("1040"))]
jpid40.JourneyPatternID.unique()
```

```python
#O

# And again we see that only a tiny number of obvious anomolies do not
# It seems that BlockID contains no extra useful information and can be dropped

jpid40.groupby(['BlockID']).size()
```

```python
# Now removing all rows where StopID = 'null' 

res = res[res.StopID != 'null']
```
